```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Membaca data ekspresi gen 3D dari file CSV
data= pd.read_csv('https://raw.githubusercontent.com/hafsakhaerunisa/Data_Skripsi_Dea/main/Data_YF_Dea.csv', delimiter=';')
df=pd.DataFrame(data)
df
```

```python
# Menghapus kolom yang tidak diperlukan, seperti ID pasien atau waktu pengambilan sampel
df = df.drop(['ID_REF'], axis=1)

# Mengisi nilai yang hilang (jika ada) dengan nilai rata-rata kolom
df = df.fillna(df.mean())

# Standardisasi data dengan modul StandardScaler
scaler = StandardScaler()
df_standardized = scaler.fit_transform(df)
df_standardized
```

```python
df_std=pd.DataFrame(df_standardized)
df_std
```
